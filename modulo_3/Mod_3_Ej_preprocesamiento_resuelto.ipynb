### Ejercicio email con apply

En el dataframe de empleados generar una columna para el correo electronico  nombre . apellido @ paygroup . cincinatti . gov  usando apply, creando una funcion llamada email_desde_nombre

```python
def email_desde_nombre(nombre_completo,prefijo):
    dummy=nombre_completo.split(",")
    apellido=dummy[0].lower()
    nombre=dummy[1].split(" ")[0].lower()
    email="{}.{}@{}.cincinnati.gov".format(nombre,apellido,prefijo.lower())
    return email
    
    
```

```python
email_desde_nombre("Moschel,Brian K.","GEN")
```

```python
empleados_limpio['email_apply']=empleados_limpio.apply(lambda x: email_desde_nombre(x['name'],x['paygroup']),axis=1)
```
